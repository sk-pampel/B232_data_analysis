## Tweezer Trap Depth Conversions

```python
def powerToDepth(power,tweezerNumber):
    total_power = power # trap depth in chimera units (mW)
    tweezer_number = tweezerNumber
    conversion_factor = 2.52 #3.2 # (mW/mK)
    T_to_Hz = 20.8366176361 # (MHz/mK)
    h = 4.1357e-15 # eV s
    depth_temp = power/(conversion_factor*tweezer_number)
    depth_freq = depth_temp*T_to_Hz
    depth_energy = h*depth_freq 

    print('trap depth  = ',round(depth_temp,5),'mK','=',round(depth_freq,5),'MHz')
    return depth_temp*1e-3

def atom_temp(trap_depth):
    KelvinToJoules = 1.380648780669e-23
    w_0 = 0.7e-6 # beam waist (radius)
    T_1mK = 15e-6 # atom temp in 1 mK trap
    U = trap_depth*KelvinToJoules # depth in Joules
    U_1mK = 1e-3*KelvinToJoules # depth in Joules
    wr_1mK = np.sqrt(4*U_1mK/(mc.Rb87_M*w_0**2)) # Hz
    w_r = np.sqrt(4*U/(mc.Rb87_M*w_0**2)) # Hz 
    nbar = 1/(np.exp(mc.hbar*wr_1mK/(mc.k_B*T_1mK))-1)
    T = mc.hbar*w_r/mc.k_B *1/(np.log(1/nbar+1))
    print('T =',T)  
    print('radial trap freq = ',w_r/(2*np.pi))
    
def tweezerToAtomTemp(power,tweezerNumber):
    total_power = power # trap depth in chimera units (mW)
    tweezer_number = tweezerNumber
    conversion_factor = 2.52 #3.2 # (mW/mK)
    T_to_Hz = 20.8366176361 # (MHz/mK)
    h = 4.1357e-15 # eV s
    depth_temp = power/(conversion_factor*tweezer_number)
    depth_freq = depth_temp*T_to_Hz
    print('trap depth  = ',round(depth_temp,5),'mK','=',round(depth_freq,5),'MHz')
    KelvinToJoules = 1.380648780669e-23
    w_0 = 0.7e-6 # beam waist (radius)
    T_1mK = 15e-6 # atom temp in 1 mK trap
    U = depth_temp*1e-3*KelvinToJoules # depth in Joules
    U_1mK = 1e-3*KelvinToJoules # depth in Joules
    wr_1mK = np.sqrt(4*U_1mK/(mc.Rb87_M*w_0**2)) # Hz
    w_r = np.sqrt(4*U/(mc.Rb87_M*w_0**2)) # Hz 
    nbar = 1/(np.exp(mc.hbar*wr_1mK/(mc.k_B*T_1mK))-1)
    T_atom = mc.hbar*w_r/mc.k_B *1/(np.log(1/nbar+1))
    print('atom temp =',T_atom)  

    print('radial trap freq = ',w_r/(2*np.pi))
```

```python
from fractions import Fraction

def decimal_to_fraction(decimal, tolerance=0.001):
    if abs(decimal - round(decimal)) < tolerance:
        # If the decimal is very close to an integer, return the integer as a fraction.
        return Fraction(int(round(decimal)))
    else:
        # Use a continued fraction expansion to approximate the decimal as a fraction.
        return Fraction.from_float(decimal).limit_denominator()

# Example usage:
decimal_value = 0.75
fraction = decimal_to_fraction(decimal_value)
print(f"The nearest fraction to {decimal_value} is {fraction}")
```

## Tweezer Volume and $\beta$ 

$V=\left(\frac{4 \pi k_b T}{m\left(\omega_r^2 \omega_a\right)^{2 / 3}}\right)^{3 / 2}$

$\beta = \Gamma V N(N-1)$

```python
def betaCalc(gamma,trap_depth,T,N): 
    KelvinToJoules = 1.380648780669e-23
    lmbda = 850e-9
    w_0 = 0.7e-6 # beam waist (radius)
    U = trap_depth*KelvinToJoules # depth in Joules
    w_r = np.sqrt(4*U/(mc.Rb87_M*w_0**2))# Hz
    w_a = w_r / (np.sqrt(2)*w_0*np.pi/lmbda) # Hz
    # w_r = 135000*(2*np.pi)
    # w_a = 25000*(2*np.pi)
    w = (w_r**2 * w_a)**(1/3)
    V = ((2*np.pi*mc.k_B * T/(mc.Rb87_M*w**2))**(3/2))*1e6 # cm^3
    beta = N*(N-1)*gamma*1e3*2*np.sqrt(2)*V # cm^3 s^-1
    print('beta=',beta,'w_r=',w_r/(2*np.pi),'w_a=',w_a/(2*np.pi))
```

```python
browayes = betaCalc(4.5,.5e-3,20-6,3)
b232 = betaCalc(4.12,.85e-3,200e-6,2)
b232 = betaCalc(15,.85e-3,30e-6,2)
```
