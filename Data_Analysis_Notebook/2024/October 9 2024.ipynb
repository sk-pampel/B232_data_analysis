```python
def calculate_and_plot_probability_distribution(averaged_image, num_bins=100):
    """
    Calculate and plot the normalized probability distribution of noise counts for the entire image,
    with a logarithmic scale for the y-axis.
    
    Parameters:
        averaged_image (2D numpy array): The input averaged image array.
        num_bins (int): Number of bins to use for the histogram.
    """
    # Flatten the image to a 1D array to combine all pixel intensities
    pixel_values = averaged_image.flatten()
    
    # Calculate the histogram for the pixel values
    counts, bin_edges = np.histogram(pixel_values, bins=num_bins)
    
    # Normalize the histogram to get a probability distribution (so that the sum equals 1)
    probability = counts / np.sum(counts)
    
    # Plot the probability distribution with a logarithmic y-axis
    plt.plot(bin_edges[:-1], probability, drawstyle='steps-post')
    plt.yscale('log')
    
    # Configure plot
    plt.title('Probability Distribution of Camera Counts')
    plt.xlabel('Camera counts')
    plt.ylabel('Probability')
    plt.grid()
    plt.show()
    
    return probability, bin_edges



# Calculate and plot the normalized probability distribution
probability, bin_edges = calculate_and_plot_probability_distribution(rawData[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_combined_histogram(rawData, num_bins=100):
    """
    Calculate the combined histogram from a list of images and plot the probability distribution.
    
    Parameters:
        rawData (list of 2D numpy arrays): The input list containing 1000 images.
        num_bins (int): Number of bins to use for the histogram.
    """
    # Initialize an array to accumulate histogram counts
    combined_counts = np.zeros(num_bins)
    bin_edges = None
    
    # Iterate over each image in the rawData list
    for image in rawData:
        # Flatten the image to a 1D array to combine all pixel intensities
        pixel_values = image.flatten()
        
        # Calculate the histogram for the current image
        counts, bin_edges = np.histogram(pixel_values, bins=num_bins, range=(0, 1000))
        
        # Accumulate the histogram counts
        combined_counts += counts
    
    # Normalize the combined histogram to get the probability distribution
    total_pixels = np.sum(combined_counts)  # Total number of pixel values across all images
    probability = combined_counts / total_pixels  # Normalize to get probabilities
    
    # Plot the probability distribution with a logarithmic y-axis
    plt.plot(bin_edges[:-1], probability, drawstyle='steps-post')
    plt.yscale('log')
    
    # Configure plot
    plt.title('Combined Probability Distribution of Camera Counts')
    plt.xlabel('Camera counts')
    plt.ylabel('Probability')
    plt.grid()
    plt.show()
    
    return probability, bin_edges

# Assuming rawData is a list containing 1000 2D numpy arrays (images)
# Example usage
probability, bin_edges = calculate_combined_histogram(rawData)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def exponential_function(x, A, tau):
    """Exponential function for fitting."""
    return A * np.exp(-x / tau)

def gaussian_function(x, A, mu, sigma):
    """Gaussian function for fitting."""
    return A * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

def calculate_combined_histogram_and_fit(rawData, num_bins=100, fit_start=200, gaussian_center=88, gaussian_range=20):
    """
    Calculate the combined histogram from a list of images, plot the probability distribution,
    fit the exponential tail of the distribution, and fit a Gaussian to the initial peak.
    
    Parameters:
        rawData (list of 2D numpy arrays): The input list containing 1000 images.
        num_bins (int): Number of bins to use for the histogram.
        fit_start (int): The bin index to start the exponential fit from (tail region).
        gaussian_center (int): The center of the initial peak for Gaussian fitting.
        gaussian_range (int): The range around the peak center to use for fitting the Gaussian.
    """
    # Initialize an array to accumulate histogram counts
    combined_counts = np.zeros(num_bins)
    bin_edges = None
    
    # Iterate over each image in the rawData list
    for image in rawData:
        # Flatten the image to a 1D array to combine all pixel intensities
        pixel_values = image.flatten()
        
        # Calculate the histogram for the current image
        counts, bin_edges = np.histogram(pixel_values, bins=num_bins, range=(0, 1000))
        
        # Accumulate the histogram counts
        combined_counts += counts
    
    # Normalize the combined histogram to get the probability distribution
    total_pixels = np.sum(combined_counts)  # Total number of pixel values across all images
    probability = combined_counts / total_pixels  # Normalize to get probabilities
    
    # Calculate bin centers
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    
    # Fit the Gaussian peak near the specified center
    gaussian_fit_indices = (bin_centers >= (gaussian_center - gaussian_range)) & (bin_centers <= (gaussian_center + gaussian_range))
    popt_gaussian, pcov_gaussian = curve_fit(
        gaussian_function, bin_centers[gaussian_fit_indices], probability[gaussian_fit_indices], 
        p0=(np.max(probability[gaussian_fit_indices]), gaussian_center, 10)
    )
    A_gaussian, mu_gaussian, sigma_gaussian = popt_gaussian
    
    # Fit the exponential tail of the distribution
    tail_indices = bin_centers >= fit_start  # Use bins from fit_start to the end for fitting
    popt_exponential, pcov_exponential = curve_fit(
        exponential_function, bin_centers[tail_indices], probability[tail_indices], 
        p0=(1, 100)
    )
    A_exp, tau_exp = popt_exponential
    
    # Plot the probability distribution with a logarithmic y-axis
    plt.plot(bin_centers, probability, drawstyle='steps-post', label='Data')
    plt.plot(bin_centers[gaussian_fit_indices], gaussian_function(bin_centers[gaussian_fit_indices], *popt_gaussian), 'g--', label=f'Gaussian fit: A={A_gaussian:.2e}, mu={mu_gaussian:.2f}, sigma={sigma_gaussian:.2f}')
    plt.plot(bin_centers[tail_indices], exponential_function(bin_centers[tail_indices], *popt_exponential), 'r--', label=f'Exponential fit: A={A_exp:.2e}, tau={tau_exp:.2f}')
    plt.yscale('log')
    
    # Configure plot
    plt.title('Combined Probability Distribution of Camera Counts')
    plt.xlabel('Camera counts')
    plt.ylabel('Probability')
    plt.grid()
    plt.legend()
    plt.show()
    
    return probability, bin_edges, A_gaussian, mu_gaussian, sigma_gaussian, A_exp, tau_exp

# Assuming rawData is a list containing 1000 2D numpy arrays (images)
# Example usage
probability, bin_edges, A_gaussian, mu_gaussian, sigma_gaussian, A_exp, tau_exp = calculate_combined_histogram_and_fit(rawData)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def gaussian_function(x, A, mu, sigma):
    """Gaussian function for fitting."""
    return A * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

def fit_gaussian_to_each_column(rawData, num_bins=100, gaussian_center=88, gaussian_range=20):
    """
    Fit a Gaussian separately for each column across all images, and then average the fitted parameters.
    
    Parameters:
        rawData (list of 2D numpy arrays): The input list containing 1000 images.
        num_bins (int): Number of bins to use for the histogram.
        gaussian_center (int): The center of the initial peak for Gaussian fitting.
        gaussian_range (int): The range around the peak center to use for fitting the Gaussian.
    """
    # Get the shape of the images
    num_columns = rawData[0].shape[1]
    
    # Storage for Gaussian fit parameters for each column
    gaussian_params = []

    # Iterate over each column
    for col in range(num_columns):
        # Initialize an array to accumulate histogram counts for this column
        combined_counts = np.zeros(num_bins)
        bin_edges = None
        
        # Iterate over each image in the rawData list
        for image in rawData:
            # Extract the current column and flatten it
            pixel_values = image[:, col]
            
            # Calculate the histogram for the current column
            counts, bin_edges = np.histogram(pixel_values, bins=num_bins, range=(0, 1000))
            
            # Accumulate the histogram counts
            combined_counts += counts
        
        # Normalize the combined histogram to get the probability distribution
        total_pixels = np.sum(combined_counts)  # Total number of pixel values across all images for this column
        probability = combined_counts / total_pixels  # Normalize to get probabilities
        
        # Calculate bin centers
        bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
        
        # Fit the Gaussian peak near the specified center
        gaussian_fit_indices = (bin_centers >= (gaussian_center - gaussian_range)) & (bin_centers <= (gaussian_center + gaussian_range))
        try:
            popt_gaussian, _ = curve_fit(
                gaussian_function, bin_centers[gaussian_fit_indices], probability[gaussian_fit_indices], 
                p0=(np.max(probability[gaussian_fit_indices]), gaussian_center, 10)
            )
            # Store the fitted parameters (A, mu, sigma)
            gaussian_params.append(popt_gaussian)
        except RuntimeError:
            # If the fit fails, append NaNs
            gaussian_params.append([np.nan, np.nan, np.nan])
    
    # Convert the list of parameters to a numpy array for easier manipulation
    gaussian_params = np.array(gaussian_params)
    
    # Average the Gaussian fit parameters across all columns, ignoring NaNs
    A_avg = np.nanmean(gaussian_params[:, 0])
    mu_avg = np.nanmean(gaussian_params[:, 1])
    sigma_avg = np.nanmean(gaussian_params[:, 2])
    
    # Plotting the averaged Gaussian fit for reference
    x_values = np.linspace(gaussian_center - 3 * gaussian_range, gaussian_center + 3 * gaussian_range, 500)
    plt.plot(x_values, gaussian_function(x_values, A_avg, mu_avg, sigma_avg), 'g--', label=f'Averaged Gaussian fit: A={A_avg:.2e}, mu={mu_avg:.2f}, sigma={sigma_avg:.2f}')
    plt.title('Averaged Gaussian Fit Over All Columns')
    plt.xlabel('Camera counts')
    plt.ylabel('Probability')
    plt.grid()
    plt.legend()
    plt.show()
    
    return A_avg, mu_avg, sigma_avg

# Assuming rawData is a list containing 1000 2D numpy arrays (images)
# Example usage
A_avg, mu_avg, sigma_avg = fit_gaussian_to_each_column(rawData)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def exponential_function(x, A, tau):
    """Exponential function for fitting."""
    return A * np.exp(-x / tau)

def fit_exponential_to_each_column(rawData, num_bins=100, fit_start=200):
    """
    Fit an exponential decay separately for each column across all images, and then average the fitted parameters.
    
    Parameters:
        rawData (list of 2D numpy arrays): The input list containing 1000 images.
        num_bins (int): Number of bins to use for the histogram.
        fit_start (int): The bin index to start the exponential fit from (tail region).
    """
    # Get the shape of the images
    num_columns = rawData[0].shape[1]
    
    # Storage for exponential fit parameters for each column
    exponential_params = []

    # Iterate over each column
    for col in range(num_columns):
        # Initialize an array to accumulate histogram counts for this column
        combined_counts = np.zeros(num_bins)
        bin_edges = None
        
        # Iterate over each image in the rawData list
        for image in rawData:
            # Extract the current column and flatten it
            pixel_values = image[:, col]
            
            # Calculate the histogram for the current column
            counts, bin_edges = np.histogram(pixel_values, bins=num_bins, range=(0, 1000))
            
            # Accumulate the histogram counts
            combined_counts += counts
        
        # Normalize the combined histogram to get the probability distribution
        total_pixels = np.sum(combined_counts)  # Total number of pixel values across all images for this column
        probability = combined_counts / total_pixels  # Normalize to get probabilities
        
        # Calculate bin centers
        bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
        
        # Fit the exponential tail starting from the specified bin index
        tail_indices = bin_centers >= fit_start  # Use bins from fit_start to the end for fitting
        try:
            popt_exponential, _ = curve_fit(
                exponential_function, bin_centers[tail_indices], probability[tail_indices], 
                p0=(np.max(probability[tail_indices]), 100)
            )
            # Store the fitted parameters (A, tau)
            exponential_params.append(popt_exponential)
        except RuntimeError:
            # If the fit fails, append NaNs
            exponential_params.append([np.nan, np.nan])
    
    # Convert the list of parameters to a numpy array for easier manipulation
    exponential_params = np.array(exponential_params)
    
    # Average the exponential fit parameters across all columns, ignoring NaNs
    A_avg = np.nanmean(exponential_params[:, 0])
    tau_avg = np.nanmean(exponential_params[:, 1])
    
    # Plotting the averaged exponential fit for reference
    x_values = np.linspace(fit_start, 1000, 500)
    plt.plot(x_values, exponential_function(x_values, A_avg, tau_avg), 'r--', label=f'Averaged Exponential fit: A={A_avg:.2e}, tau={tau_avg:.2f}')
    plt.title('Averaged Exponential Fit Over All Columns')
    plt.xlabel('Camera counts')
    plt.ylabel('Probability')
    plt.grid()
    plt.legend()
    plt.yscale('log')
    plt.show()
    
    return A_avg, tau_avg

# Assuming rawData is a list containing 1000 2D numpy arrays (images)
# Example usage
A_avg, tau_avg = fit_exponential_to_each_column(rawData)
```
