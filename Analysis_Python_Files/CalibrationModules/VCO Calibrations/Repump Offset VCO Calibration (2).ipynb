# Repump Offset VCO Calibration

I believe the VCO is a POS-300

MOT frequency = 219 MHz @ dac voltage -0.6

```python
from matplotlib.pyplot import *
%matplotlib inline
```

```python
# dac 21
offVoltage = [-0.6, 0, 1, 2, 3, 4, 3.5, -1, -2, -3, -4, -5, -3.5, -3.4, -3.3, 
           -3.2, 3.25, 3.4, 3.6, 3.7, 2.5, 2.75]
# (MHz)
offFrequency = [219.3, 204.5, 182, 162.5, 141.0, 107.3, 126.0, 229.5, 254.3, 278.3, 286.5, 286.5, 286.5, 286.5, 286.5, 
             283.5, 135.0, 130.5, 121.5, 117.8, 153, 147.8]
# dbm, 1 dc block on front of RF source, nothing else.
offPower = [-24.35 ,-24.7, -25.8, -27.05, -29.25, -52.8, -32.9, -24.15, -23.85, -23.65, -23.7, -23.7, -23.8, -23.8, -23.9, 
            -23.7, -30.4, -31.6, -34.9, -37.9, -27.9, -28.5]
```

```python
plot(offVoltage, offFrequency, 'o')
title('VCO WAS OFF')
# nonlinearity occurs at ~3.4V or so
figure()
title('VCO WAS OFF')
plot(offFrequency, offPower, 'ro')
figure()
title('VCO WAS OFF')
plot(offVoltage, offPower, 'go');
```

Realized after I did this that I did the calibration with the vco in the off setting.

```python
dacVoltage = [-0.6, 3.0, 4.0, 3.5, 3.6, 3.7, 3.8, 3.9, 1.5, 0.75, -1, -2, -3, -4, -3.5, -3.2, -3.3]
freq = [219.0, 141.0, 107.3, 126, 121.5, 117.8, 114.8, 111.0, 171.8, 187.5, 228.8, 253.5, 278.3, 286.5, 286.5, 282.8, 285.8]
power = [13.23, 13.18, -5.4, 12.2, 11.47, 9.51, 5.75, 0.45, 13.18, 13.15, 12.99, 12.8, 12.63, 12.52, 12.52, 12.58, 12.56]
```
