# Calculation of the Movre-Pischler Potentials for an ${}^{87}\mathrm{Rb}$ dimer

The Movre-Pischler potentials are the energies of a diatomic molecule when including Born-Oppenhiemer and fine structure interactions. <br>
To calculate the energies, we need to find the following matrix elements as a function of internuclear distance R:

$\left \langle L', |\Lambda|', S', |\Sigma|', \chi',\kappa' \right| \hat H_{BO}+\hat H_{FS} \left|L ,|\Lambda|, S, |\Sigma|, \chi, \kappa \right \rangle  
=  E_{BO}  + \left \langle |L', |\Lambda|', S', |\Sigma|', \chi',\kappa' \right | A_{FS}(l_a \cdot s_a  + l_b \cdot s_b) \left|L ,|\Lambda|, S, |\Sigma|, \chi, \kappa \right \rangle $

where $E_{BO}=\frac{2C_3}{R^{3}[(1 + |\Lambda|)!]^2} (-1)^{|\Lambda|+S}\chi $ are the Born-Oppenheimer energies.

The fine structure Hamiltonian $ \hat H_{FS}$ is not diagonal in the molecular basis, so we need to perform a frame transformation to  <br> $|j_a,\omega_a,l_a,s_a \rangle |j_b,\omega_b,l_b,s_b \rangle$ . This Hamiltonian will yield off-diagonal matrix elements in the molecular basis.

### generate BO states  
$ l_a,l_b,s_a,s_b  \rightarrow |L, |\Lambda|, S, \chi, \kappa_{BO} \rangle $

takes single atom l and s values to generate BO states <br> 

Using the seperate atom values l and s, we first generate the core Born-Oppenheimer states needed to calculate $E_{BO}$ . <br>
For two Rb atoms in an S + P molecular state, we can take $l_a=0,l_b=1$ and $s_a=s_b=1/2$ <br>
$\chi= \pm 1$ is the inversion symmetry quantum number and $\kappa_{BO} = 1$ is the reflection symmetry quantum number. <br>
Only states with $\Lambda=0$ have well-defined reflection symmetry.

```python
def create_BO_states(l_a,l_b,s_a,s_b):
    BO_states = []
    L = l_a+l_b
    kappa_BO = 1
    for Lmda in np.arange(0,np.max(L)+1,1):
        for S in np.arange(0,s_a+s_b+1,1):
            for chi in [1,-1]:
                if Lmda == 0:
                    bostates = {'L':L,'|Lambda|':Lmda,'S':int(S),'chi':chi,'kappa_BO':kappa_BO}
                else:
                    bostates = {'L':L,'|Lambda|':Lmda,'S':int(S),'chi':chi,'kappa_BO':''}
                BO_states.append(bostates)
    return BO_states
```

```python
print('\033[1m' + 'BO States'+ '\033[0m')
BO_states = create_BO_states(0,1,1/2,1/2) 
for state in BO_states:
    print(state)
```

### generate FS-expanded BO states 

$ |L, |\Lambda|, S, \chi,\kappa_{BO} \rangle \rightarrow |L, |\Lambda|, S, |\Sigma|, |\Omega|,\chi, \kappa_{BO}, \kappa_{FS} \rangle $

expand BO states to include states split by fine structure symmetries.

We need to expand the bare BO states to include quantum numbers relevant for fine structure interactions.<br> 
This includes $|\Sigma|$ and  $|\Omega|$, the projections of S and J along the internuclear axis . <br>
However, by including $|\Sigma|$ and $|\Omega|$, the reflection symmetry of the state is now different from the BO state, <br> so we must introduce $\kappa_{FS}$.  

```python
def expand_BO_states_FS(BO_states):
    expanded_BO_states = []
    for state in BO_states:
        for Sigma in np.arange(-state['S'], state['S']+1,1):
            Omega = Sigma+state['|Lambda|']
            if state['|Lambda|'] == 1 and Sigma == -1: # 3Pi_g states --> Omega=0
                for kappa_FS in [-1,1]:
                    exp_bostates = {'L':state['L'],'|Lambda|':state['|Lambda|'],'S':state['S'],'chi':state['chi'],
                                           'kappa_BO':state['kappa_BO'],'|Sigma|':abs(Sigma), '|Omega|':abs(Omega), 'kappa_FS':kappa_FS }  
                    expanded_BO_states.append(exp_bostates)
            elif Sigma >=0: # Lambda=[0,1]  Sigma = [0,1]  Omega = [0,1,2]       
                kappa_FS = (-1)**(state['L']-state['|Lambda|']+state['S']-Sigma)
                exp_bostates = {'L':state['L'],'|Lambda|':state['|Lambda|'],'S':state['S'],'chi':state['chi'],
                                       'kappa_BO':state['kappa_BO'],'|Sigma|':abs(Sigma), '|Omega|':abs(Omega), 'kappa_FS':kappa_FS }            
                expanded_BO_states.append(exp_bostates)
    return expanded_BO_states
```

```python
expanded_BO_states = expand_BO_states_FS(BO_states)
print('\033[1m' + 'Expanded BO States'+ '\033[0m')
print('number of expanded BO states: ',len(expanded_BO_states))
for state in expanded_BO_states:
    print(state)
```

### Group expanded BO states by fine structure symmetry 

The expanded BO states can be grouped according to their fine structure symmetries labelled as $\Omega^{\kappa_{FS}}_{g/u}$. <br> Each group comprises a block diagonal matrix in the total Hamiltonian. Since states fo different symmetry do not mix, <br>
we can eventually diagonalize each of these matrices to find the energies of states with specific symmetry.

```python
def group_BO_states_by_FS_symmetry(expanded_BO_states):
    FS_symmetry_BO_states = {}
    for Omega in [0,1,2]:
        for state in expanded_BO_states:
            chi_label = 'g' if state['chi']==1 else 'u'
            if state['|Omega|']==Omega: 
                if Omega == 0:
                    parityString = '+' if state['kappa_FS'] == 1 else '-'
                else:
                    parityString = ''
                symmetryLabel = str(Omega) + parityString + chi_label
                if symmetryLabel not in FS_symmetry_BO_states:
                    FS_symmetry_BO_states[symmetryLabel] = []
                FS_symmetry_BO_states[symmetryLabel].append(state)
    return FS_symmetry_BO_states
```

```python
FS_symmetry_BO_states = group_BO_states_by_FS_symmetry(expanded_BO_states)
print('\033[1m' +'expanded BO basis grouped by fine structure symmetry '+'\033[0m')
for label, state in FS_symmetry_BO_states.items():
    print('\033[1m'+label+'\033[0m',len(state),'states',state)
```

### generate case A basis (no absolute values)

$|L,S \rangle \rightarrow |L, \Lambda, S, \Sigma, \Omega,\chi \rangle $

takes coupled atom L and S values to create expanded BO basis including negative values of $\Lambda, \Sigma, \Omega$.
Note that $\kappa$ is not part of these states.  <br>  

The fine structure interactions break the degeneracy of $\Lambda, \Sigma$ and $ \Omega$ in the expanded BO states, <br>
so we must convert the expanded BO states into "case A Born-Oppenhiemer" states $ \ | L, \Lambda, S, \Sigma, \Omega,\chi \rangle $ <br>
This function creates the case A BO basis. These are the basis states for the total Hamiltonian. 

```python
def create_case_A_basis(L_vals, S_vals, statenum=False):
    case_A_basis = []
    for chi in [1,-1]: 
        for L in L_vals: 
            for Lambda in np.arange(-L,L+1,1): 
                for S in S_vals: 
                    for Sigma in np.arange(-S, S+1,1): 
                        state = {'L':L, 'Lambda': Lambda, 'S':S,
                                 'Sigma':Sigma, 'Omega':Sigma+Lambda,'chi': chi}
                        case_A_basis.append(state)

    if statenum == True:
        for state_num, state in enumerate(case_A_basis):
            state.update({'state_num':state_num})
    return case_A_basis       
```

### generate seperate atom ls and j bases

$ l_a,l_b,s_a,s_b  \rightarrow |l_a,m_{l_a},s_a,m_{s_a},l_b,m_{l_b},s_b,m_{s_b} \rangle$ <br>
$l_a,l_b,s_a,s_b  \rightarrow |j_a,m_{j_a},l_a,s_a,j_b,m_{j_b},l_b,s_b \rangle$

takes single atom l and s values to generate all possible states in seperate atom ls and j basis.

In order to transform the BO states to the j basis, we must transform the BO states to the ls basis and then from the ls to the j basis. <br>
The easiest way to accomplish this is to make a transformation matrix of Clebsh-Gordon coefficients for every possible separate atom ls and j state <br>
(given our single atom l and s values). This function generates the necessary seperate atom ls and j basis states.

```python
def create_seperate_atom_bases(l_vals, s_vals) :   
    ls_basis, j_basis = [],[]    
    for l_a in l_vals:
        for l_b in l_vals:
            for s_a in s_vals:
                for s_b in s_vals:
                    for j_a in set(np.arange(abs(s_a - l_a), s_a + l_a + 1, 1)):
                        for j_b in set(np.arange(abs(s_b - l_b), s_b + l_b + 1, 1)):
                            for m_j_a in np.arange(-j_a,j_a+1,1):
                                for m_j_b in np.arange(-j_b,j_b+1,1):
                                    j_2atom_states = {'j_a':j_a,'m_j_a':m_j_a,'l_a':l_a,'s_a':s_a,'j_b':j_b,'m_j_b':m_j_b,'l_b':l_b,'s_b':s_b}
                                    j_basis.append(j_2atom_states)
                    for m_l_a in np.arange(-l_a,l_a+1,1):
                        for m_l_b in np.arange(-l_b,l_b+1,1):
                             for m_s_a in np.arange(-s_a, s_a+1,1):
                                for m_s_b in np.arange(-s_b,s_b+1,1):
                                    ls_2atom_states = {'l_a':l_a,'m_l_a':m_l_a,'s_a':s_a,'m_s_a':m_s_a,'l_b':l_b,'m_l_b':m_l_b, 's_b':s_b,'m_s_b':m_s_b}
                                    ls_basis.append(ls_2atom_states)


    return ls_basis, j_basis 
```

### Expanded BO to case A basis operator

$| L, \Lambda, S, \Sigma, \Omega,\chi \rangle \langle L, |\Lambda|, S, |\Sigma|, |\Omega|,\chi, \kappa_{BO}, \kappa_{FS} |$

transforms expanded BO states into case A states. From this transformation we can identify $\Sigma$ and $\Pi$ basis states needed to construct MP block diagonal matrices. <br> Since $\Pi$ states have indefinite reflection symmetry, this transformation generates superpositions of $\Pi$ states that have definite parity. This function returns the index and sign of the matching case A states contained in the case A basis <br>
which can be used to construct the superposition basis states of the total Hamiltonian.

for states ${|L, |\Lambda|, S, |\Sigma|, |\Omega|,\chi, \kappa_{BO}, \kappa_{FS} \rangle}_p$  and ${|L, \Lambda, S, \Sigma, \Omega,\chi \rangle}_q$ <kb> where p = 0,1,2,...,15 and q = 0,1,2,...,23


${|L, |\Lambda|, S, |\Sigma|, |\Omega|,\chi, \kappa_{BO}, \kappa_{FS} \rangle}_p= {|L, \Lambda, S, \Sigma, \Omega,\chi \rangle}_p +\kappa_{FS} \delta_{\pm \Lambda, \pm 1} {|L, -\Lambda, S, -\Sigma, -\Omega , \chi \rangle}_p $ for each permuation of p and q

```python
def expanded_BO_to_case_A_op(state, case_A_BO_basis, indexes=False):

    # identifies Sigma (Lambda = 0) states 
    if state['|Lambda|'] == 0 and state['|Sigma|'] == 0:

        Sigma_caseA_state = {}
        for key in state.keys():
            if key in ['kappa_BO', 'kappa_FS']: # remove kappa
                pass
            elif key[0] != "|": # if there are no absolute value signs
                Sigma_caseA_state[key] = state[key]
            else:
                Sigma_caseA_state[key[1:-1]] = 0 # set Lambda,Sigma,Omega = 0
        if indexes:
            # print([case_A_basis.index(Sigma_caseA_state)], [1])
            return [case_A_basis.index(Sigma_caseA_state)], [1],[Sigma_caseA_state]
        
         
        # return converted state as ket |L,Lambda,S,I_BO,Sigma,Omega>
        print('|'+''.join([str(val) for key, val in Sigma_caseA_state.items()])+'>')
        return '|'+''.join([str(val) for key, val in Sigma_caseA_state.items()])+'>'

    # identifies Pi (Lambda = 1) superposition states 
    else:
        Pi_caseA_state_1, Pi_caseA_state_2 = {}, {}
        # print(state)
        for key in state.keys():
            if key in ['kappa_BO', 'kappa_FS']:
                pass
            elif key[0] != "|":
                Pi_caseA_state_1[key] = state[key]
                Pi_caseA_state_2[key] = state[key]
            elif key == "|Lambda|":
                Pi_caseA_state_1[key[1:-1]] = state[key]
                Pi_caseA_state_2[key[1:-1]] = -state[key] # flipping sign of Lambda due to action of kappa
            elif key == "|Sigma|":
                Pi_caseA_state_1[key[1:-1]] = state['|Omega|'] - Pi_caseA_state_1['Lambda']
                Pi_caseA_state_2[key[1:-1]] = -state['|Omega|'] - Pi_caseA_state_2['Lambda']
            elif key == "|Omega|":
                Pi_caseA_state_1['Omega'] = state['|Omega|']
                Pi_caseA_state_2['Omega'] = -state['|Omega|']
        sign = '+' if state['kappa_FS'] == 1 else '-'
        #sign = '+' if state['kappa_BO'] == 1 else '-'
        # print('sym1',stateMostlySym1,'sym2',stateMostlySym2) 
        if indexes:
            # print([case_A_basis.index(Pi_caseA_state_1), case_A_basis.index(Pi_caseA_state_2)], [1,1 if sign == "+" else -1])
            return [case_A_basis.index(Pi_caseA_state_1), case_A_basis.index(Pi_caseA_state_2)], [1,1 if sign == "+" else -1],[Pi_caseA_state_1,Pi_caseA_state_2]
                # return state numbers (indices) for superposition states to build basis (Pi states), sign of each superposotion state
        print('|'+''.join([str(val) for key, val in Pi_caseA_state_1.items()])+'>'+sign+'|'+''.join([str(val) for key, val in Pi_caseA_state_2.items()])+'>')
        return '|'+''.join([str(val) for key, val in Pi_caseA_state_1.items()])+'>'+sign+'|'+''.join([str(val) for key, val in Pi_caseA_state_2.items()])+'>'
    

```

### case A to ls basis transformation 

#### case A to ls transformation matrix

$|l_{a}, m_{l_a}, s_{a}, m_{s_a}, l_{b}, m_{l_b}, s_{b}, m_{s_b} \rangle \langle L, \Lambda, S, \Sigma, \Omega ,\chi   |\\ $ this operator is an intermediate step in the transformation between the expanded BO basis and j basis

for states ${|L, \Lambda, S, \Sigma, \Omega,\chi \rangle}_p$  and ${|l_{a}, m_{l_a}, s_{a}, m_{s_a}, l_{b}, m_{l_b}, s_{b}, m_{s_b}  \rangle}_q$ <kb> where p = 0,1,2,...,23 and q = 0,1,2,...,63


we need to find $ _q\langle {l_a, m_{l_a}, s_a, m_{s_a},l_b, m_{l_b},s_b, m_{s_b}|}{L, \Lambda, S, \Sigma, \Omega,\chi \rangle}_p=(-1)^{S}\chi \ C_{{(l_a m_{l_a} l_b m_{l_b})}_p}^{{(\Lambda \lambda_a)}_q}C_{{(s_a m_{s_a} s_b m_{s_b})}_p}^{{(S \Sigma)}_q}$ for permuations of p and q

let $ \ \ \ C_{(ls)_q}^{(BO)_p}=(-1)^{S_p}\chi_p \ C_{{(l_a m_{l_a} l_b m_{l_b})}_q}^{{(\Lambda \lambda_a)}_p}C_{{(s_a m_{s_a} s_b m_{s_b})}_q}^{{(S \Sigma)}_p}$ 

The case A BO to ls operator is a 63x23 matrix of the form


$\left(\begin{array}{ccccc}
C_{(ls)_0}^{(BO)_0} & C_{(ls)_0}^{(BO)_1} & C_{(ls)_0}^{(BO)_2} & \cdots & C_{(ls)_0}^{(BO)_{23}} \\
C_{(ls)_1}^{(BO)_0} & C_{(ls)_1}^{(BO)_1} & C_{(ls)_1}^{(BO)_2} & \cdots & C_{(ls)_1}^{(BO)_{23}}\\
C_{(ls)_2}^{(BO)_0} & C_{(ls)_2}^{(BO)_1} & C_{(ls)_2}^{(BO)_2} & \cdots & C_{(ls)_2}^{(BO)_{23}}\\
\vdots & \vdots & \vdots & \ddots \\
C_{(ls)_{63}}^{BO_{0}} & C_{(ls)_{63}}^{BO_{1}} & C_{(ls)_{63}}^{BO_{2}} & \cdots & C_{(ls)_{63}}^{BO_{23}}
\end{array}\right)$

```python
def case_A_BO_to_ls_op(case_A_BO_basis,ls_basis): 
    op = np.zeros((len(case_A_BO_basis), len(ls_basis)))
    for case_A_num, case_A_state in enumerate(case_A_BO_basis):
        for ls_num, ls_state in enumerate(ls_basis): 
            # if (case_A_state['S'] not in np.arange(abs(ls_state['s_a']-ls_state['s_b']), ls_state['s_a'] + ls_state['s_b']+1,1) # if S not in [0,1]
            #     or case_A_state['Lambda'] != ls_state['m_l_a']+ls_state['m_l_b'] # remove lsi 2 atom states that don't satisfy definition of Lambda = lambda_a + lambda_b
            #     or case_A_state['Sigma'] != ls_state['m_s_a'] + ls_state['m_s_b'] # " " Sigma
            #     or ls_state['l_a'] == ls_state['l_b']): # remove if l_a = l_b
            #     continue
            op[case_A_num,ls_num] = ((-1)**case_A_state['S']*case_A_state['chi']*float(CG(ls_state['l_a'], ls_state['m_l_a'], ls_state['l_b'], ls_state['m_l_b'], case_A_state['L'], case_A_state['Lambda']).doit())
                                    *float(CG(ls_state['s_a'], ls_state['m_s_a'], ls_state['s_b'], ls_state['m_s_b'], case_A_state['S'], case_A_state['Sigma']).doit()))
    return op           
```

```python
def case_A_BO_to_ls_op(case_A_basis,ls_basis): 
    op = np.zeros((len(case_A_basis), len(ls_basis)))
    for case_A_num, case_A_state in enumerate(case_A_basis):
        for ls_num, ls_state in enumerate(ls_basis): 
            if (case_A_state['S'] not in np.arange(abs(ls_state['s_a']-ls_state['s_b']), ls_state['s_a'] + ls_state['s_b']+1,1) 
                or case_A_state['Lambda'] != ls_state['m_l_a']+ls_state['m_l_b'] 
                or case_A_state['Sigma'] != ls_state['m_s_a'] + ls_state['m_s_b'] 
                or ls_state['l_a'] == ls_state['l_b']): 
                continue
            op[case_A_num,ls_num] = (((-1)**case_A_state['S']*case_A_state['chi'])**ls_state['l_a']*float(CG(ls_state['l_a'], ls_state['m_l_a'], ls_state['l_b'], ls_state['m_l_b'], case_A_state['L'], case_A_state['Lambda']).doit())
                                    *float(CG(ls_state['s_a'], ls_state['m_s_a'], ls_state['s_b'], ls_state['m_s_b'], case_A_state['S'], case_A_state['Sigma']).doit()))
        op[case_A_num,:] /= np.linalg.norm(op[case_A_num,:])
    return op           
```

```python
def get_column_state(case_A_basis, basis_states):
    column_state = [[0] for _ in range(len(case_A_basis))] # create list of 24 [0]'s 
    for num, state in enumerate(case_A_basis): # 24 states
        # print(num,state)
        match = True
        for qnum, val in basis_states.items(): 
            # print(qnum,val,state[qnum])
            if val != state[qnum]:
                match = False
        if match: # indicate column state where quantumNums = state
            column_state[num][0] = 1 
            # print(column_state)
            return column_state
        # returns 24 lists of length 24 containing one entry of [1] along diagonal
    raise ValueError("No Match! nums were" + str(basis_states))


# for state in case_A_basis:  
#     gcs = get_column_state(case_A_basis,state)
#     print(gcs)
```

#### case A to ls grouped by FS symmetry

```python
def case_A_to_ls_grouped(FS_sym_BO_state, ls_basis, case_A_basis):
    final_matrix = np.zeros((len(FS_sym_BO_state),len(ls_basis)))
    case_A_to_ls_matrix = case_A_BO_to_ls_op(case_A_basis,ls_basis)
    
    for sym_BO_state_num, sym_BO_state in enumerate(FS_sym_BO_state):  # iterate through expanded BO states grouped by FS symetry
        indexes, signs, state = expanded_BO_to_case_A_op(sym_BO_state, case_A_basis, indexes=True) # for each expanded BO state, find index and sign of case A transformed state
        new_state = 0 

        # for superpostion states
        if sym_BO_state['|Omega|'] != 0: # group indices and signs together
            indexes = [indexes[0]] #[5, 4] [1, -1] --> [5,1] 
            signs = [signs[0]] # [5, 4] [1, -1] --> [4,-1]
        for index, sign in zip(indexes,signs): # group index and sign in list of tuples [(index 1,sign 1), (index 2,sign 2)]
         
            case_A_state = case_A_basis[index] # find case A states corresponding to index
            column_state_matrix = get_column_state(case_A_basis, case_A_state) # find corresponding column state for case A row state
            new_state += sign * case_A_to_ls_matrix.T@ column_state_matrix      
        final_matrix[sym_BO_state_num, :] = new_state[:,0] / np.sqrt(len(indexes));   

    return final_matrix


case_A_basis = create_case_A_basis([1],[0,1]) 
for fs_sym_label, fs_sym_bo_state in FS_symmetry_BO_states.items():
    print('\033[1m' +fs_sym_label+'\033[0m')
    case_A_to_ls = case_A_to_ls_grouped(fs_sym_bo_state, ls_basis, case_A_basis).T
    # np.set_printoptions(threshold=np.inf)
    # print(case_A_to_ls)
```

### create ls to j basis transformation matrix

$|l_a, m_{l_a}, s_a, m_{s_a},l_b, m_{l_b},s_b, m_{s_b} \rangle \langle j_a, m_{j_a}, l_a,s_a,j_b, m_{j_b},l_b,s_b|$

this function creates the transformation matrix to from the ls to j basis.

for states ${|l_a, m_{l_a}, s_a, m_{s_a},l_b, m_{l_b},s_b, m_{s_b} \rangle}_p$  and ${|j_a, m_{j_a}, l_a,s_a,j_b, m_{j_b},l_b,s_b \rangle}_q$ <kb> where p = q = 0,1,2,...,63

we need to find $ _q\langle {j_a, m_{j_a}, l_a,s_a,j_b, m_{j_b},l_b,s_b|}{l_a, m_{l_a}, s_a, m_{s_a},l_b, m_{l_b},s_b, m_{s_b} \rangle}_p=C_{{(l_a m_{l_a} s_a m_{s_a})}_p}^{{(j_a m_{j_a})}_q}C_{{(l_b m_{l_b} s_b m_{s_b})}_p}^{{(j_b m_{j_b})}_q}$ for each permuation of states p and q

let $ \ \ \ C_{(ls)_p}^{j_q}=C_{{(l_a m_{l_a} s_a m_{s_a})}_p}^{{(j_a m_{j_a})}_q}C_{{(l_b m_{l_b} s_b m_{s_b})}_p}^{{(j_b m_{j_b})}_q}$ 

The ls to j operator is a matrix of the form


$\left(\begin{array}{ccccc}
C_{(ls)_0}^{j_0} & C_{(ls)_1}^{j_0} & C_{(ls)_2}^{j_0} & \cdots & C_{(ls)_{63}}^{j_0} \\
C_{(ls)_0}^{j_1} & C_{(ls)_1}^{j_1} & C_{(ls)_2}^{j_1} & \cdots & C_{(ls)_{63}}^{j_1}\\
C_{(ls)_0}^{j_2} & C_{(ls)_1}^{j_2} & C_{(ls)_2}^{j_2} & \cdots & C_{(ls)_{63}}^{j_2}\\
\vdots & \vdots & \vdots & \ddots \\
C_{(ls)_{63}}^{j_{0}} & C_{(ls)_{63}}^{j_{1}} & C_{(ls)_{63}}^{j_{2}} & \cdots & C_{(ls)_{63}}^{j_{63}}
\end{array}\right)$

```python
def ls_to_j_op(ls_basis, j_basis):
    op = np.zeros((len(ls_basis),len(j_basis))) 
    for ls_num, ls_state in enumerate(ls_basis): 
        for j_num, j_state in enumerate(j_basis):
            op[j_num,ls_num] = 1
            if ls_state['l_a'] != j_state['l_a'] or ls_state['l_b'] != j_state['l_b'] or ls_state['s_a'] != j_state['s_a'] or ls_state['s_b'] != j_state['s_b']:
                op[j_num,ls_num] *= 0
            else:
                op[j_num,ls_num] *= (float(CG(ls_state['l_a'], ls_state['m_l_a'], ls_state['s_a'], ls_state['m_s_a'], j_state['j_a'], j_state['m_j_a']).doit())
                                       *float(CG(ls_state['l_b'], ls_state['m_l_b'], ls_state['s_b'], ls_state['m_s_b'], j_state['j_b'], j_state['m_j_b']).doit()))
    return op
```

### create fine structure Hamiltonian 

$\langle j_a, \omega_{a},l_a, s_a, j_b, \omega_{b}, l_b, s_b |\hat H_{FS} |j_a, \omega_{a},l_a, s_a, j_b, \omega_{b}, l_b, s_b \rangle $

This function creates a diagonal matrix of eigenvalues for all possible j states.

Operates on j basis states to give eigenvalues $ \ \ \hat H_{FS} |j_a, \omega_{a},l_a, s_a, j_b, \omega_{b}, l_b, s_b  \rangle =E_{FS}|j_a, \omega_{a},l_a, s_a, j_b, \omega_{b}, l_b, s_b \rangle$

$ E_{FS} = A_{FS}(l_a \cdot s_a  + l_b \cdot s_b) = \frac{2}{3} \Delta (l_a \cdot s_a  + l_b \cdot s_b)$ <kb>
$ =\frac{\Delta}{3} (j_a^2-l_a^2-s_a^2)+(j_b^2-l_b^2-s_b^2) $

$= \frac{\Delta}{3}[j_a(j_a+1) - l_a(l_a+1)-s_a(s_a+1)+j_b(j_b+1) - l_b(l_b+1)-s_b(s_b+1)] $

Values are contained in a diagonal 64 x 64 matrix of the form


$\left(\begin{array}{ccccc}
E_{0,0} & 0 & 0 & \cdots & 0 \\
0 & E_{1,1} & 0 & \cdots & 0\\
0 & 0 & E_{2,2} & \cdots & 0\\
\vdots & \vdots & \vdots & \ddots \\
0 & 0 & 0 & \cdots & E_{63,63}
\end{array}\right)$

```python
# delta = 4.7197635548748796
delta = 1 
def create_H_FS(j_basis, delta):
    H_FS = np.zeros((len(j_basis), len(j_basis)))
    for state_num, state in enumerate(j_basis): 
        H_FS[state_num,state_num] = delta/3 * (state['j_a']*(state['j_a']+1)-state['l_a']
                *(state['l_a']+1)-state['s_a']*(state['s_a']+1) + state['j_b']*(state['j_b']+1)-state['l_b']
                *(state['l_b']+1)-state['s_b']*(state['s_b']+1))
    return H_FS
```

### create BO Hamiltonian

$E_{BO}=\frac{2C_3}{R^{3}[(1 + |\Lambda|)!]^2} (-1)^{|\Lambda|+S}\chi $ 

```python

zero_minus_g = [{'L': 1, '|Lambda|': 0, 'S': 1, 'I_BO': 'g', 'kappa_BO': 1, '|Sigma|': 0, 'kappa_FS': 1, '|Omega|': 0, '|Iota|': 0, 'I': 0, 'i_a': 0, 'i_b': 0, '|Phi|': 0, 'kappa_HFS': 1}]

def get_H_BO(C3, Rv, bo_basis):
    # returns the Born Oppenheimer Hamiltonian
    # expects states to be a list of lists where the low level list has the values of L, Lambda, S,Sigma, and I_BO.
    #H_BO = np.array([[0.0 for _ in bo_basis] for _ in bo_basis])
    H_BO = np.zeros((len(bo_basis),len(bo_basis)))
    lambdaKey = "|Lambda|" if "|Lambda|" in bo_basis[0] else "Lambda"
    # the matrix is diagonal.
    for num, state in enumerate(bo_basis):
        I_BO = 1 if state["I_BO"]=="g" else -1
        pv = (-1)**(state["S"]+I_BO)
        L_ = state["L"]
        H_BO[num,num] = -pv*(3*state[lambdaKey]**2-L_*(L_+1))/Rv**3 * C3
    return H_BO


hbo = get_H_BO(C3_val, 20*a_0 , zero_minus_g)
print((hbo)*1e21)
```

```python
C3_val = 5.94492796981645e-48
a_0 = 0.52917721067e-10
h = 6.6260700e-34
zero_minus_g = [{'L': 1, '|Lambda|': 0, 'S': 1, 'chi': 1, 'kappa_BO': 1, '|Sigma|': 0, '|Omega|': 0, 'kappa_FS': 1}]
def create_H_BO(C3, R, bo_basis):
    H_BO = np.zeros((len(bo_basis),len(bo_basis)))
    for num, state in enumerate(bo_basis):
        H_BO[num,num] = 2*(C3/R**3)*(-1)**(state['|Lambda|']+state['S'])*state['chi']/math.factorial((1+state['|Lambda|']))**2
    return H_BO


hbo = create_H_BO(C3_val,20*a_0 , zero_minus_g)
print((hbo)*1e21)
```
