# Calculation of the Movre-Pischler Potentials with hyperfine interactions for an ${}^{87}\mathrm{Rb}$ dimer

The following calculations give the energies of a diatomic molecule when including Born-Oppenhiemer, fine structure, and hyperfine strucutre interactions. <br>


### Summary of theory

To calculate the energies, we need to find the following matrix elements as a function of internuclear distance R for each possible state:

$\langle \Psi' | \hat H_{BO}+ \hat H_{FS} + \hat H_{HFS} | \Psi \rangle = E_{BO} + E_{FS} + \langle \Psi' | A_{hfs}( \vec{i}_a \cdot \vec{j}_a +\vec{i}_b \cdot \vec{j}_b ) | \Psi \rangle $

where $|\Psi \rangle = \frac{1}{\sqrt{2-\delta_{|\Lambda|,0}}}\left( |L, \Lambda, S, \Sigma, \chi, \sigma_v, I, \mathcal{I},\sigma_{v}' \rangle +  (-1)^{\Lambda+S-\Sigma+I-\mathcal{I}} \delta_{|\Lambda|,1} |L, -\Lambda, S, -\Sigma, \chi, \sigma_v, I, -\mathcal{I},\sigma_{v}'\rangle \right)$ are the Born-Oppenheimer basis states

The hyperfine structure Hamiltonian $ \hat H_{HFS}$ is not diagonal in the Born-Oppenheimer basis, so we do a frame transformation to the $|f_a \phi_a (j_a i_a) f_b \phi_b (j_b i_b) \rangle$ basis.

 The desired calculation is then $ \hat U^{\dagger} \hat H_{HFS} \hat U$  where the matrix elements of $\hat U$ are given by

$ \langle l_{a} \lambda_{a} s_{a} \sigma_{a} l_{b} \lambda_{b} s_{b} \sigma_{b}| L, \Lambda, S, \Sigma, \chi, \sigma_v, I, \mathcal{I},\sigma_{v}' \rangle = $ $(-1)^{(S+\chi+1)^{l_a}} \ C_{l_a \lambda_{a} l_b \lambda_{b}}^{L \Lambda}C_{s_a \sigma_{a} s_b \sigma_{b}}^{S \Sigma} C_{i_a \phi_{a} i_b \phi_{b}}^{I \mathcal{I} }$

$ \langle {j_a \omega_{a} l_as_ai_a\phi_{a}j_b \omega_{b}l_bs_bi_b\phi_{b}|}{l_a \lambda_{a} s_a \sigma_{a}i_a\phi_{a}l_b \lambda_{b}s_b \sigma_{b}i_b\phi_{b} \rangle}_p=C_{{l_a \lambda_{a} s_a \sigma_{a}}}^{{j_a \omega_{a}}}C_{{l_b \lambda_{b} s_b \sigma_{s_b}}}^{{j_b \omega_{b}}}$ for the lsi to ji transformation and

$ \langle {f_a \phi_{a} j_a l_a  s_ai_af_b \phi_{b} j_b l_b  s_bi_b|}{j_a \omega_{a} l_as_ai_a \iota_{a}j_b \omega_{b}l_bs_bi_biota_{b} \rangle}=C_{{j_a \omega_{a} i_a \iota_{a}}}^{{f_a \phi_{a}}}C_{{j_b \omega_{b} i_b \iota_{b}}}^{{f_b \phi_{b}}}$ for the ji to f transformation

Letting $ \ \ \mathscr{D}_{l_a l_b s_a s_b i_a i_b}^{L \Lambda S \Sigma I \mathcal{I}}=(-1)^{(S+\chi+I)^{l_a}} \ C_{l_a \lambda_{a} l_b \lambda_{b}}^{L \Lambda}C_{s_a \sigma_{a} s_b \sigma_{b}}^{S \Sigma} C_{i_a \phi_{a} i_b \phi_{b}}^{I \mathcal{I} }\ \  \ $              $ \ \ \ \ \ \ \ \mathscr{D}_{l_as_al_bs_b}^{j_a \omega_a j_b \omega_b}=C_{l_a \lambda_{a} s_a \sigma_{a}}^{j_a \omega_{a}}C_{l_b \lambda_{b} s_b \sigma_{b}}^{j_b \omega_{b}}$   $ \ \ \ \ \ \ \ \ \ \mathscr{D}_{j_a i_a j_b i_b}^{f_a \phi_a f_b \phi_b}=C_{{j_a \omega_{a} i_a \iota_{a}}}^{{f_a \phi_{a}}}C_{{j_b \omega_{b} i_b \iota_{b}}}^{{f_b \phi_{b}}}$

$ \hat U^{\dagger} \hat H_{FS} \hat U
= \left( {\left(\begin{array}{ccccc}
\mathscr{D}_{l_a l_b s_a s_b i_a i_b}^{L \Lambda S \Sigma I \mathcal{I}}\cdots \\
\vdots \ \ \ \ \ \ \ \ddots 
\end{array}\right)} {\left(\begin{array}{ccccc}
\mathscr{D}_{l_as_al_bs_b}^{j_a \omega_a j_b \omega_b}\cdots \\
\vdots \ \ \ \ \ \ \ \ddots 
\end{array}\right)} {\left(\begin{array}{ccccc}
\mathscr{D}_{j_a i_a j_b i_b}^{f_a \phi_a f_b \phi_b} \cdots \\
\vdots \ \ \ \ \ \ \ \ddots 
\end{array}\right)} \right)^{\dagger}\left(\begin{array}{ccccc}
E_{HFS} & 0 & \cdots \\
0 &  \ddots  \\
\vdots & & \ddots  
\end{array}\right)\left( {\left(\begin{array}{ccccc}
\mathscr{D}_{l_a l_b s_a s_b i_a i_b}^{L \Lambda S \Sigma I \mathcal{I}}\cdots \\
\vdots \ \  \ \ \ \ \ \ddots 
\end{array}\right)} {\left(\begin{array}{ccccc}
\mathscr{D}_{l_as_al_bs_b}^{j_a \omega_a j_b \omega_b}\cdots \\
\vdots \ \ \ \ \ \ \ \ddots 
\end{array}\right)} {\left(\begin{array}{ccccc}
\mathscr{D}_{j_a i_a j_b i_b}^{f_a \phi_a f_b \phi_b} \cdots \\
\vdots \ \ \ \ \ \ \ \ddots 
\end{array}\right)} \right) $




The fine structure Hamiltonian is given by 

$ \hat H_{hfs} = \sum_{f_a \phi_{a} f_b \phi_{b}}|f_a \phi_{a} (l_a s_a i_a) f_b \phi_{b} (l_b s_b i_b) \rangle E_{hfs} \langle f_a \phi_{a} (l_a s_a i_a) f_b \phi_{b} (l_b s_b i_b) | $ 

Since the hyperfine splittings are different for D1 and D2 lines, we must calculate the energies for each atom seperately and add them together. The energy for a single atom is 

$E_{\alpha}=A_{hfs} \vec{i}_{\alpha} \cdot \vec{j}_{\alpha}+\delta_{j_{\alpha},\frac{3}{2}}B_{hfs} \frac{3(\vec{i}_{\alpha}  \cdot \vec{j}_{\alpha} )^2+\frac{3}{2} \vec{i}_{\alpha}  \cdot \vec{j}_{\alpha}-i_{\alpha}(i_{\alpha}+1) j_{\alpha}(j_{\alpha}+1)}{2 i_{\alpha}(2 i_{\alpha}-1) j_{\alpha}(2 j_{\alpha}-1)} \approx \frac{A_{hfs}}{2} \left(f_{\alpha}^2-j_{\alpha}^2-i_{\alpha}^2\right) =\frac{A_{hfs}}{2} \left[f_{\alpha}(f_{\alpha}+1) - j_{\alpha}(j_{\alpha}+1)-i_{\alpha}(i_{\alpha}+1) \right]$ 


where $A_{hfs}$ is the magnetic dipole constant that determines the splitting between adjacent hyperfine states for D1 and D1

The energies, $E_{hfs} = E_a + E_b $ are contained in a diagonal 1024 x 1024 matrix of the form


$\left(\begin{array}{ccccc}
E_{0,0} & 0 & 0 & \cdots & 0 \\
0 & E_{1,1} & 0 & \cdots & 0\\
0 & 0 & E_{2,2} & \cdots & 0\\
\vdots & \vdots & \vdots & \ddots \\
0 & 0 & 0 & \cdots & E_{1023,1023}
\end{array}\right)$

## $\Phi=0$

```python
ival = 0
caseAMostlySymHfs = mol.create_unsym_BO_HF_basis([1],[0,1],[ival]) # 
caseASymHfs = mol.create_sym_BO_HF_states(0,1,1/2,1/2,ival,ival)
symmetrizedBases,mostlySymmetrizedBases = mol.get_phi_bases(caseASymHfs,caseAMostlySymHfs)
lsiBasis, jiBasis, fBasis, lsiBasis2Atom, jiBasis2Atom, fBasis2Atom = mol.createAtomicBases([0,1], [1/2], [ival])
fsH = mol.create_H_FS(jiBasis2Atom)* 2 * E_FS_5P12_5P32 / 3
hfsH = mol.create_H_HFS(fBasis2Atom, E_HFS_5P12_F1F2, E_HFS_5S12_F1F2, cs.Rb87_5P32_ToF3,cs.Rb87_5P32_ToF2,cs.Rb87_5P32_ToF1,cs.Rb87_5P32_ToF0) * cs.h
lsiToJi_Op = mol.create_lsiToJi_Op(lsiBasis,jiBasis) #########
jiToF_Op = mol.create_jiToF_Op(jiBasis, fBasis) #####
```

```python
H_fs_phi, H_hfs_phi = {}, {}
caseastates = []

# for label, base in normalized_symbases.items():
for label, base in symmetrizedBases.items():
    # IPython.display.clear_output(wait=True)
    # print('Working on states of symmetry ' + label +". " + str(len(base)) + " states")
    fsStates, hfsStates = [],[]
    for num, state in enumerate(base):
        # misc.reportProgress(num, len(base))
        indexes, signs = mol.caseASymHfsToMostlySym(state, caseAMostlySymHfs, indexes=True)
        symHfsState, symFsState = 0, 0
        # construct the correct superposition which will preserve the symmetry of the given base. 
        # This should be only either 1 or 2 iterations in this loop.
        for index, sign in zip(indexes,signs):
            caseAState = caseAMostlySymHfs[index] 
            I_BO = 0 if caseAState["chi"] == "g" else 1
            nuclearNums = [caseAState['I'], caseAState['Iota'],ival,ival]

            oalNums = (caseAState["L"], caseAState["Lambda"], 1, 0) # oal = "orbital angular momentum"
            spinNums = (caseAState["S"], caseAState["Sigma"], 1/2, 1/2)

            symHfsState += sign * mol.caseAToAtomic( oalNums, spinNums,nuclearNums, I_BO, lsiBasis, basisChange=jiToF_Op@lsiToJi_Op )
            symFsState += sign * mol.caseAToAtomic( oalNums, spinNums,nuclearNums, I_BO, lsiBasis, basisChange=lsiToJi_Op )
            # caseastates.append(symFsState)

        fsStates.append(symFsState / np.sqrt(len(indexes)));
        hfsStates.append(symHfsState / np.sqrt(len(indexes)))
        # caseastates.append(hfsStates)

    # print(symFsState)
    H_fs_phi[label] = mol.convertH_toCaseABasis( fsStates, fsH, offset = -1/2* 2 * E_FS_5P12_5P32 / 3)
    H_hfs_phi[label] = mol.convertH_toCaseABasis( hfsStates, hfsH, offset=0)   
    # print(fsStates)
```

```python
for label, matrix in H_hfs_phi.items():
    print('\033[1m' +label+'\033[0m \n', matrix*1e21)
```

## $\Phi=1-5$

```python
# for label, base in normalized_mostlysymbases.items():
for label, base in mostlySymmetrizedBases.items():
    # IPython.display.clear_output(wait=True)
    # print('Working on states of symmetry ' + label +". " + str(len(base)) + " states")
    fsStates, hfsStates = [[], []]
    states = []
    for num, state in enumerate(base[:]):
        misc.reportProgress(num, len(base))
        I_BO = mol.g if state["chi"] == "g" else mol.u
        nuclearNums = [state['I'], state['Iota'],ival,ival]
        oalNums = (state["L"], state["Lambda"], 1, 0)
        spinNums = (state["S"], state["Sigma"], 1/2, 1/2)
        fsStates.append(mol.caseAToAtomic( oalNums, spinNums, nuclearNums, I_BO, lsiBasis, basisChange=lsiToJi_Op ));
        hfsStates.append(mol.caseAToAtomic( oalNums, spinNums, nuclearNums, I_BO, lsiBasis, basisChange=jiToF_Op@lsiToJi_Op ))
        states.append(spinNums)
    H_fs_phi[label] = mol.convertH_toCaseABasis( fsStates, fsH, offset=-1/2 * 2 * E_FS_5P12_5P32 / 3)
    H_hfs_phi[label] = mol.convertH_toCaseABasis( hfsStates, hfsH, offset=0)

    # print(states)    
```

```python
allBases = {**symmetrizedBases , **mostlySymmetrizedBases}
points = 2000
Rvals = np.linspace( 1*20*cs.a0, 100*20*cs.a0, points )
energies = {}
for label, base in allBases.items():
    IPython.display.clear_output(wait=True)
    print('\nLabel:', label, '\nCalcualting Eigenvalues at every R')
    energies[label] = np.zeros((len(Rvals), len(base)))
    for num, rv in enumerate(Rvals):
        misc.reportProgress(num, len(Rvals))
        # print(base)
        totalEnergy = mol.get_H_BO(C_3_val, rv, base).astype(float) + H_fs_phi[label] + H_hfs_phi[label]
        eigVals, eigVectors = np.linalg.eigh(totalEnergy)
        energies[label][num] = eigVals
```

```python
def getLabelColorAndLinestyle(label):
    phiColors=['purple','blue','green', 'orange','red', '#FF69B4']       
    if label[0] == '0':
        if label[-1] == '1':
            color = 'black'
        elif label[-1] == '2':
            color = '#800080'
    else:
        color = phiColors[int(label[0])]
    linestyle = 'dashed' if label[1] == 'b' else 'solid'
    return color, linestyle
```

```python
def normalize_values(dictionary):
    """
    Normalize numeric values in a dictionary to integers.
    """
    normalized_dict = {}
    for key, value in dictionary.items():
        if isinstance(value, (int, float)):
            normalized_dict[key] = int(value)
        else:
            normalized_dict[key] = value
    return normalized_dict

def normalize_keys(dictionary):
    """
    Normalize the keys of a dictionary.
    """
    normalized_dict = {}
    for key, value in dictionary.items():
        if key == 'kappa_FS':
            normalized_dict['sigma_v'] = value
        elif key == 'I_BO':
            normalized_dict['chi'] = value
        elif key == 'kappa_HFS':
            normalized_dict['sigma_HFS'] = value
        elif key not in ['i_a', 'i_b', 'kappa_BO']:  # Ignore 'i_a' and 'i_b'
            normalized_dict[key] = value

    return normalized_dict

def normalize_list_of_dicts(list_of_dicts):
    """
    Normalize a list of dictionaries.
    """
    normalized_list = []
    for dictionary in list_of_dicts:
        normalized_dict = normalize_values(dictionary)
        normalized_dict = normalize_keys(normalized_dict)
        normalized_list.append(normalized_dict)
    return normalized_list

# Define the original list of dictionary lists

# Normalize each sublist of dictionaries individually
normalized_mostlysymbases= {key: normalize_list_of_dicts(value) for key, value in markmostlysym.items()}
```

```python
def normalize_values(dictionary):
    """
    Normalize numeric values in a dictionary to integers.
    """
    normalized_dict = {}
    for key, value in dictionary.items():
        if isinstance(value, (int, float)):
            normalized_dict[key] = int(value)
        else:
            normalized_dict[key] = value
    return normalized_dict

def normalize_keys(dictionary):
    """
    Normalize the keys of a dictionary.
    """
    normalized_dict = {}
    for key, value in dictionary.items():
        if key == 'kappa_FS':
            normalized_dict['sigma_v'] = value
        elif key == 'I_BO':
            normalized_dict['chi'] = value
        elif key == 'kappa_HFS':
            normalized_dict['sigma_HFS'] = value
        elif key not in ['i_a', 'i_b', 'kappa_BO']:  # Ignore 'i_a' and 'i_b'
            normalized_dict[key] = value

    return normalized_dict

# Normalize the dictionaries in the lists
list1 = stevenunsym
list2 = markunsym
normalized_list1 = [normalize_values(item) for item in list1]
normalized_list2 = [normalize_values(item) for item in list2]

# Normalize keys for list of dictionaries 1
normalized_list1 = [normalize_keys(d) for d in normalized_list1]

# Normalize keys for list of dictionaries 2
normalized_list2 = [normalize_keys(d) for d in normalized_list2]

# Sort the lists of dictionaries
sorted_list1 = sorted(normalized_list1, key=lambda x: sorted(x.items()))
sorted_list2 = sorted(normalized_list2, key=lambda x: sorted(x.items()))

# Compare the sorted lists of dictionaries
if sorted_list1 == sorted_list2:
    print("Both lists contain the same dictionaries.")
else:
    print("The dictionaries in the lists are different.")
```
